```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, poisson, norm
```

```python
# Set up a figure with subplots
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle("Analysis of Binomial Distributions", fontsize=16)
# ------------------- BINOMIAL DISTRIBUTION -------------------
n, p = 10, 0.4
x_binom = np.arange(0, n+2)
pmf_binom = binom.pmf(x_binom, n, p)
cdf_binom = binom.cdf(x_binom, n, p)
samples_binom = binom.rvs(n, p, size=1000)

# PMF
axs[0].bar(x_binom, pmf_binom, color='red')
axs[0].set_title("Binomial PMF (n=10, p=0.4)")
# CDF
axs[1].step(x_binom, cdf_binom, where='mid', color='skyblue')
axs[1].set_title("Binomial CDF")
# Histogram of Samples
axs[2].hist(samples_binom, bins=np.arange(n+2)-0.5, density=True, color='navy', alpha=0.7)
axs[2].set_title("Binomial Samples Histogram")
```

```python
# Set up a figure with subplots
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle("Poisson Distributions", fontsize=16)
# ------------------- POISSON DISTRIBUTION -------------------
mu = 4
x_poisson = np.arange(0, 15)
pmf_poisson = poisson.pmf(x_poisson, mu)
cdf_poisson = poisson.cdf(x_poisson, mu)
samples_poisson = poisson.rvs(mu, size=1000)

# PMF
axs[0].bar(x_poisson, pmf_poisson, color='lightgreen')
axs[0].set_title("Poisson PMF (μ=4)")
# CDF
axs[1].step(x_poisson, cdf_poisson, where='mid', color='green')
axs[1].set_title("Poisson CDF")
# Histogram of Samples
axs[2].hist(samples_poisson, bins=np.arange(16)-0.5, density=True, color='darkgreen', alpha=0.7)
axs[2].set_title("Poisson Samples Histogram")
```

```python
# Set up a figure with subplots
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle("Normal Distributions", fontsize=16)
# ------------------- NORMAL DISTRIBUTION -------------------
mu_norm, sigma = 0, 1
x_norm = np.linspace(-4, 4, 1000)
pdf_norm = norm.pdf(x_norm, mu_norm, sigma)
cdf_norm = norm.cdf(x_norm, mu_norm, sigma)
samples_norm = norm.rvs(mu_norm, sigma, size=1000)

# PDF
axs[0].plot(x_norm, pdf_norm, color='salmon')
axs[0].set_title("Normal PDF (μ=0, σ=1)")
# CDF
axs[1].plot(x_norm, cdf_norm, color='red')
axs[1].set_title("Normal CDF")
# Histogram of Samples
axs[2].hist(samples_norm, bins=30, density=True, color='darkred', alpha=0.7)
axs[2].set_title("Normal Samples Histogram")

# # Layout and show
# for row in axs:
#   for a in row:
#     a.grid(True)
# plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python


# ------------------- BASIC STATISTICAL ANALYSIS -------------------
print("Binomial Distribution:")
print(f"Mean = {np.mean(samples_binom):.2f}, Variance = {np.var(samples_binom):.2f}")
print("\nPoisson Distribution:")
print(f"Mean = {np.mean(samples_poisson):.2f}, Variance = {np.var(samples_poisson):.2f}")
print("\nNormal Distribution:")
print(f"Mean = {np.mean(samples_norm):.2f}, Standard Deviation = {np.std(samples_norm):.2f}")
```
